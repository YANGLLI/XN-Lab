```python
#features_to_use = ["BUSINESS","STATE"]

master_bus_st = pd.read_sql_query('SELECT business, state, zip FROM master', engine)
business_bus_st = pd.read_sql_query('SELECT businessname, mailingstatecode, mailingzip FROM businesslicense', engine)
```

```python
print(master_bus_st)
```

```python
master_co = master_bus_st[(master_bus_st.state=='CO')]
business_co = business_bus_st[business_bus_st.mailingstatecode=='CO']
print(master_co)
print(business_co)
```

```python
length = len(business_co)
business_co['distance'] = pd.Series(-1, index=business_co.index)
business_co['master'] = pd.Series("", index=business_co.index)
business_co['master_zip'] = pd.Series(0, index=business_co.index)
print(business_co)
```

```python
import distance

for index,row in business_co.iterrows():
    min_distance = 10000
    master_name = "xxx"
    master_zip = 0
    business_name = row['businessname']
    
    for index1, row1 in master_co.iterrows():
        business = row1['business']
        n = distance.levenshtein(business_name.lower(), business.lower())
        zipcode = row1['zip']
        if(n < min_distance):
            min_distance = n
            master_name = business
            master_zip = zipcode
    business_co.loc[index, 'distance'] = min_distance
    business_co.loc[index, 'master'] = master_name
    business_co.loc[index, 'master_zip'] = master_zip
```

```python
business_co[business_co['distance'] < 10].to_csv("compare.csv")
```
